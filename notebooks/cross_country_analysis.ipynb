```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load all 5 cleaned datasets
files = ['ethiopia_cleaned.csv', 'kenya_cleaned.csv', 'sudan_cleaned.csv', 
         'nigeria_cleaned.csv', 'tanzania_cleaned.csv']

data_frames = [pd.read_csv(f'../data/{f}') for f in files]
df_master = pd.concat(data_frames, ignore_index=True)

# 2. Analytical Visualization: Regional Temperature Distribution
plt.figure(figsize=(12, 6))
sns.violinplot(data=df_master, x='Country', y='T2M', inner="quart", palette="muted")
plt.title('Regional Temperature Distribution: Task 3 Comparison')
plt.ylabel('Mean Temperature (°C)')
plt.show()
```
