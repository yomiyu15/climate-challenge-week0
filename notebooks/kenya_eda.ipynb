```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Load the Kenya dataset 
df = pd.read_csv('../data/kenya.csv')

# 2. Feature Engineering
df['Country'] = 'Kenya'
df['Date'] = pd.to_datetime(df['YEAR'] * 1000 + df['DOY'], format='%Y%j')
df['Month'] = df['Date'].dt.month

# 3. Scientific Data Cleaning
# Replace NASA's -999 placeholder with NaN
df.replace(-999, np.nan, inplace=True)

# 4. Data Quality & Missing Value Report
print(f"Duplicate rows: {df.duplicated().sum()}")
missing_report = pd.DataFrame({
    'Missing Values': df.isna().sum(),
    'Percentage (%)': (df.isna().sum() / len(df)) * 100
})
display(missing_report)

# 5. Outlier Detection (|Z| > 3)
cols_to_check = ['T2M', 'T2M_MAX', 'T2M_MIN', 'PRECTOTCORR', 'RH2M', 'WS2M']
z_scores = np.abs(stats.zscore(df[cols_to_check], nan_policy='omit'))
print(f"\nExtreme weather days identified: {(z_scores > 3).any(axis=1).sum()}")

# 6. Save Cleaned Data for Task 3
df.to_csv('../data/kenya_cleaned.csv', index=False)
print("\nSuccess: kenya_cleaned.csv saved!")
```
